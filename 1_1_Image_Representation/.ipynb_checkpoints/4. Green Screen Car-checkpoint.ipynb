# Color Threshold, Green Screen

### Import resources

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2

%matplotlib inline
```

### Read in and display the image

```python
# Read in the image
image = mpimg.imread('images/car_green_screen.jpg')

# Print out the image dimensions (height, width, and depth (color))
print('Image dimensions:', image.shape)
```

```python
# Display the image
plt.imshow(image)
```

### Define the color threshold

```python
## TODO: Define our color selection boundaries in RGB values
#exclude pixels that are between lower and upper bounds
lower_green = np.array([0,200,0]) 
upper_green = np.array([195,255,205])
```

### Create a mask

```python
# Define the masked area
mask = cv2.inRange(image, lower_green, upper_green)

# Vizualize the mask
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the car show through
masked_image = np.copy(image)

masked_image[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```

### Mask and add a background image

```python
# Load in a background image, and convert it to RGB 
background_image = mpimg.imread('images/sky.jpg')
print("background_image.shape: {}".format(background_image.shape))

## TODO: Crop it or resize the background to be the right size (450x660)
dim = (660, 450)
background_image_resized = cv2.resize(background_image, dim, interpolation = cv2.INTER_AREA)
print("background_image_resized.shape: {}".format(background_image_resized.shape))

## TODO: Mask the cropped background so that the car area is blocked
# Hint mask the opposite area of the previous image
masked_background_image = np.copy(background_image_resized)
masked_background_image[mask == 0] = [0, 0, 0]

## TODO: Display the background and make sure 
plt.imshow(masked_background_image)
```

### Create a complete image

```python
## TODO: Add the two images together to create a complete image!
# complete_image = masked_image + crop_background

final_image = masked_image + masked_background_image

plt.imshow(final_image)
```
